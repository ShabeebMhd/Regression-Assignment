```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset = dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,criterion='absolute_error',max_features='log2',random_state=0)
regressor=regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_Random_Forest.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_Random_Forest.sav", 'rb'))
result=loaded_model.predict([[50,35,2,1,0]])
```
